1. Difference Between a Function and a Method in Python
- Function: A block of reusable code that is independent and can be called using its name. Example: print(), len().
Method: A function that is associated with an object and is called on it.
Example: "hello".upper() (the upper() method acts on a string object).

2. Function Arguments and Parameters in Python
- Parameters: The variables defined in the function definition (e.g., def func(a, b): here a and b are parameters).
Arguments: The actual values passed to the function when it is called (e.g., func(5, 10) here 5 and 10 are arguments).

3. Ways to Define and Call a Function in Python
Define: Using the def keyword:
- def greet(name):
  return f"Hello, {name}!"
Call: By using the function name followed by parentheses:
print(greet("Alice"))

4. Purpose of the return Statement
The return statement is used to send a result back to the caller of the function. It exits the function and optionally passes back a value.
Example:


def add(a, b):
return a + b
result = add(3, 4)  # result is 7

5. Iterators vs. Iterables
- Iterable: An object that can be looped over (e.g., lists, strings).
Iterator: An object that keeps state and produces the next value in an iterable when next() is called.
Difference: Iterables use iter() to get an iterator, while iterators can be consumed one value at a time.

6. Generators in Python
- Generators are functions that yield values one at a time using the yield keyword, instead of returning them all at once. Example:

def my_gen():
    yield 1
    yield 2
    yield 3
gen = my_gen()

7. Advantages of Generators
- Memory efficient: They generate values on demand.
Simplify code: No need to manage state or use loops explicitly.
Useful for handling large data streams.

8. Lambda Functions
- A lambda function is an anonymous function defined using the lambda keyword.
Example:

square = lambda x: x**2
print(square(5))  # Outputs: 25
Used in scenarios requiring short, simple functions.

9. Purpose of map()
- The map() function applies a given function to all items in an iterable.
Example:

nums = [1, 2, 3]
squared = map(lambda x: x**2, nums)
print(list(squared))  # Outputs: [1, 4, 9]

10. Difference Between map(), reduce(), and filter()
- map(): Transforms each element.
filter(): Filters elements based on a condition.
reduce(): Aggregates elements to a single value.
Example:

from functools import reduce

nums = [1, 2, 3, 4]
mapped = list(map(lambda x: x**2, nums))  # [1, 4, 9, 16]
filtered = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]
reduced = reduce(lambda x, y: x + y, nums)  # 10

11. Reduce Function Mechanism (Sum Operation)
For the list [47, 11, 42, 13], the reduce function adds numbers iteratively:

Steps:

Start with the first two elements: 47 + 11 = 58
Add the next element: 58 + 42 = 100
Add the last element: 100 + 13 = 113
Final result: 113

4. Check if a number is prime

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

primes = [n for n in range(1, 201) if is_prime(n)]

5. Fibonacci iterator class

class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        self.count += 1
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        return fib
6. Generator for powers of 2

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
7. File line-by-line generator

def file_line_reader(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
8. Lambda to sort tuples by second element

tuples_list = [(1, 5), (3, 2), (2, 8)]
sorted_tuples = sorted(tuples_list, key=lambda x: x[1])
9. Convert temperatures using map()

def celsius_to_fahrenheit(celsius_temps):
    return list(map(lambda c: c * 9 / 5 + 32, celsius_temps))

10. Remove vowels using filter()

def remove_vowels(string):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', string))
