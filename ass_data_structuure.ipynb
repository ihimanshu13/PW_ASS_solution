```python
What are data structures, and why are they important?
 - Data structures are ways to organize and store data efficiently for access, modification, and manipulation.
Importance:

Optimize algorithm performance.
Solve problems like searching and sorting efficiently.
Examples: Lists, dictionaries, stacks, queues.

                                                                  
2. Explain the difference between mutable and immutable data types with examples.
 - Mutable: Can be changed after creation. Example: Lists.

my_list = [1, 2, 3]
my_list[0] = 10  # Modifies the list
Immutable: Cannot be changed. Example: Strings.


my_string = "hello"
# my_string[0] = 'H'  # Raises an error

3. Main differences between lists and tuples in Python?
 - Lists: Mutable, dynamic size, slower.
Tuples: Immutable, faster, often used for fixed data.

4. How do dictionaries store data?
 - Dictionaries use hash tables, mapping keys to values via hashing. This ensures fast lookups.

5. Why might you use a set instead of a list?
 - Sets automatically remove duplicates.
Ideal for mathematical operations like union and intersection.

6. What is a string in Python, and how is it different from a list?
 - A string is an immutable sequence of characters, while a list is a mutable collection of elements, including strings, numbers, etc.


 7. How do tuples ensure data integrity in Python?
 - Tuples are immutable, ensuring data remains unchanged throughout program execution.

8. What is a hash table, and how does it relate to dictionaries in Python?
 - A hash table is a data structure for key-value storage with efficient lookups. Python’s dictionaries use hash tables internally.

9. Can lists contain different data types in Python?
 - Yes, lists can store elements of different data types.

mixed_list = [1, "string", 3.14, True]

     
10. Why are strings immutable in Python?
 - For performance, memory optimization, and security. Immutable strings can be reused and shared efficiently.

11. Advantages of dictionaries over lists?
 - Faster lookups (O(1)).
Store structured data using key-value pairs.

12. When is using a tuple preferable to a list?
- When the data should remain unchanged, such as coordinates or settings.

13. How do sets handle duplicate values?
- Sets automatically discard duplicates.

my_set = {1, 2, 2, 3}  # Results in {1, 2, 3}
14. How does the “in” keyword work for lists and dictionaries?
Lists: Checks if the value exists (O(n)).
Dictionaries: Checks if the key exists (O(1)).

15. Can you modify the elements of a tuple? Explain why or why not.
- No, tuples are immutable. To change a tuple, you must create a new one.

16. What is a nested dictionary? Example use case?
- A dictionary within another dictionary. Example:


nested_dict = {
    "person1": {"name": "Alice", "age": 25},
    "person2": {"name": "Bob", "age": 30}
}
Use case: Organizing hierarchical data.

17. Time complexity of accessing elements in a dictionary?
 - Average case: O(1). Worst case: O(n) (due to hash collisions).

18. When are lists preferred over dictionaries?
 - Data needs to maintain order.
No key-value relationship is required.

19. Why are dictionaries considered unordered?
 - Dictionaries don’t rely on insertion order for retrieval (before Python 3.7). This doesn’t affect retrieval speed but changes order expectations.

20. List vs Dictionary in data retrieval?

 - List: Retrieve by index or scan (O(n)).
Dictionary: Retrieve by key (O(1)).
```
