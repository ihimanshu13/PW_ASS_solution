1. What is Python, and why is it popular?
  - Python is a high-level, interpreted programming language known for its simplicity and readability. It is popular due to:

Easy Syntax: Resembles natural language, making it beginner-friendly.
Versatility: Supports multiple programming paradigms (object-oriented, procedural, functional).
Extensive Libraries: Provides libraries for various fields like web development (Django), data science (NumPy, pandas), machine learning (TensorFlow), and more.
Large Community: A strong support community for problem-solving and learning.

2. What is an interpreter in Python?
- An interpreter in Python is a program that reads and executes Python code line by line. It converts high-level Python code into machine-readable instructions at runtime.

3. What are pre-defined keywords in Python?
- Pre-defined keywords are reserved words in Python with special meanings. They define the language's syntax and structure, such as if, else, for, while, def, and class. 

4. Can keywords be used as variable names?
- No, keywords cannot be used as variable names because they have predefined meanings in Python's syntax.

5. What is mutability in Python?
- Mutability refers to whether an object's value can be changed after it is created:

Mutable objects: Their values can be modified (e.g., list, dict).
Immutable objects: Their values cannot be changed (e.g., tuple, str, int).

6. Why are lists mutable, but tuples are immutable?
- Lists are designed for dynamic operations, allowing items to be added, removed, or changed.
Tuples are immutable to ensure integrity and faster performance in scenarios where data must remain constant.

7. What is the difference between == and is operators in Python?
- Compares the values of two objects.
is: Checks whether two objects refer to the same memory location (identity).
Example:


a = [1, 2]
b = [1, 2]
print(a == b)  # True, as values are the same.
print(a is b)  # False, as they are different objects in memory.

8. What are logical operators in Python?
Logical operators are used to perform logical operations:

and: Returns True if both conditions are true.
or: Returns True if at least one condition is true.
not: Reverses the logical state of the condition.
Example:


x = 5
print(x > 3 and x < 10)  # True
print(x > 10 or x < 3)  # False
print(not(x < 10))      # False

9. What is type casting in Python?
Type casting is converting one data type into another:

Implicit casting: Performed automatically by Python (e.g., int to float).
Explicit casting: Done manually using functions like int(), float(), or str().

11. How does the elif statement work?
elif stands for "else if." It checks additional conditions if the previous if or elif statements are false. Example:


x = 10
if x > 15:
    print("Greater than 15")
elif x > 5:
    print("Between 5 and 15")  # This gets executed
else:
    print("5 or less")

    
12. What is the difference between for and while loops?
for loop: Iterates over a sequence (e.g., list, string, range) and is ideal when the number of iterations is known.
while loop: Continues to execute as long as a condition is true, suitable when the number of iterations is unknown.

14. Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is more suitable when the loop depends on a condition that is not tied to a predefined sequence. For example:


# Read user input until they enter "quit"
user_input = ""
while user_input != "quit":
    user_input = input("Enter something (type 'quit' to exit): ")
    print("You entered:", user_input)
